# Análisis de la Tasa de Cambio del Peso Colombiano (TRM)

**Notebook 01 — Exploración inicial**

Este notebook realiza la primera exploración de la **Tasa Representativa del Mercado (TRM)**, es decir, el precio del dólar (USD) expresado en pesos colombianos (COP).

**Objetivos de esta etapa:**

1. Cargar y limpiar la base de datos histórica de la TRM.
2. Verificar el rango temporal disponible y el tipo de cada variable.
3. Acotar el análisis a los **últimos 5 años (2021–2026)**, el período de mayor interés.
4. Generar una primera visualización que muestre la evolución de la TRM e identifique sus puntos máximo y mínimo.

> **Fuente de datos:** Banco de la República de Colombia — Tasa Representativa del Mercado (TRM).
> https://www.banrep.gov.co/es/estadisticas/trm

## 1) Carga y limpieza de la base

Cargamos el archivo CSV y corregimos el formato de las fechas: la columna `Periodo` viene como texto (`YYYY/MM/DD`) y la convertimos a tipo `datetime`. Esto es indispensable para poder ordenar los registros, filtrarlos por rangos de fechas y graficarlos correctamente sobre el eje temporal.

```python
import pandas as pd

df = pd.read_csv('../Data/raw/Tasa de cambio del peso colombiano.csv')

# Las fechas vienen como texto; se convierten a tipo datetime.
df["Periodo"] = pd.to_datetime(df["Periodo"], format="%Y/%m/%d")
```

### Verificación inicial

Revisamos las primeras filas, los tipos de dato y las dimensiones del conjunto para confirmar que la carga y la conversión de fechas fueron correctas. También consultamos la fecha mínima y máxima para conocer la cobertura histórica de la serie.

```python
display(df.head())
display(df.dtypes)
display(df.shape)
```

```python
display(df["Periodo"].min())
display(df["Periodo"].max())
```

## 2) Acotar el período de análisis (2021–2026)

Aunque la serie histórica arranca en 1991, este estudio se concentra en los **últimos 5 años**. Este recorte permite analizar la dinámica reciente del peso —marcada por la recuperación pos-pandemia, la coyuntura política y las presiones cambiarias globales— sin que las magnitudes de décadas anteriores distorsionen la escala del análisis.

Creamos el subconjunto `df_5` con los registros posteriores al **1 de enero de 2021**.

```python
df_5 = df[df["Periodo"] >= pd.Timestamp("2021-01-01")]
```

```python
display(df_5.head())
display(f"Cantidad de filas: {df_5.shape}")
display(f"Fecha mínima: {df_5['Periodo'].min():%Y-%m-%d}  |  Fecha máxima: {df_5['Periodo'].max():%Y-%m-%d}")
```

## 3) Visualización exploratoria

Graficamos la evolución de la TRM en el período 2021–2026 e identificamos de forma automática sus valores extremos:

- **Línea roja (vertical):** fecha en la que el dólar alcanzó su valor **máximo**.
- **Línea verde (vertical):** fecha en la que el dólar registró su valor **mínimo**.

Estos puntos sirven como referencia para el análisis por año que se desarrolla en el siguiente notebook.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 5))              # tamaño del canvas (ancho, alto)
plt.plot(df_5["Periodo"], df_5["TRM"])

# punto mas alto del periodo
idx_max = df_5["TRM"].idxmax()
fecha_max = df_5.loc[idx_max, "Periodo"]
valor_max = df_5.loc[idx_max, "TRM"]

# punto mas bajo del periodo
idx_min = df_5["TRM"].idxmin()
fecha_min = df_5.loc[idx_min, "Periodo"]
valor_min = df_5.loc[idx_min, "TRM"]

# linea en lo mas alto
plt.axvline(fecha_max, color="red", linestyle="--", alpha=0.7)   # linea VERTICAL en el maximo
plt.scatter(fecha_max, valor_max, color="red", zorder=5)

# linea en lo mas bajo
plt.axvline(fecha_min, color="green", linestyle="--", alpha=0.7)
plt.scatter(fecha_min, valor_min, color="green", zorder=5)

plt.grid(alpha=0.3)
plt.title("Evolución del Dólar (TRM)\n2021-Actualidad")
plt.xlabel("Tiempo")
plt.ylabel("Valor del Dólar (COP)")
plt.show()
```

## 4) Conclusiones de la exploración

- La serie de la TRM está **limpia y completa**: cubre desde **1991-11-27** hasta **2026-07-08** (12.643 registros), con las fechas ya en formato `datetime`. El subconjunto de análisis `df_5` (2021–2026) contiene **2.015 registros**.

- Entre 2021 y 2026 el dólar mostró una **alta volatilidad**: partió de **\$3.432,50** (01-ene-2021) y llegó a **\$3.335,50** (08-jul-2026), una variación acumulada de aproximadamente **−2,8 %**.

- El **máximo histórico** del período se alcanzó el **05-nov-2022** con **\$5.061,21**, mientras que el **mínimo** fue de **\$3.334,93** el **04-jul-2026**. Entre el pico y el valle el peso se **apreció cerca de un 34 %**.

- Este comportamiento —un fuerte ascenso hasta finales de 2022 seguido de una recuperación del peso— motiva un **análisis año por año** para relacionar los movimientos de la TRM con el contexto económico y político de cada período.
